```python
## Deleting the null rows:
df_del = df.dropna(axis=0)
df_del.shape
```

```python
X = df_del.drop([ 'Price'], axis = 1)
X.shape
```

```python
y = df_del[['Price']]
y.shape
```

```python
## Feature Selection
features = ['living area', 'number of bedrooms','number of bathrooms']
target = 'Price'  # assuming 'Price' is the target variable'

X = df[features]
y = df[target]
```

```python
from sklearn.impute import SimpleImputer ## Handle missing values
```

```python
imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=features)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() ## normalize scaling feature
X_scaled = pd.DataFrame(scaler.fit_transform(X_imputed), columns=features)
```

```python
from sklearn.model_selection import train_test_split ## train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() ##train linear regression model
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
## Evaluation metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2_score =r2_score(y_test,y_pred)
```

```python
print(f"MAE: {mae:}, RMSE: {rmse:},r2_score:{r2_score}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting Actual vs Predicted values
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, color='blue', edgecolor='w', alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.xlabel("Actual House Price")
plt.ylabel("Predicted House Price")
plt.title("Actual vs Predicted House Prices India")
plt.grid(True)
plt.tight_layout()
plt.show()
```
